```python
def cal_precision_recall_map_plot(reterived_docs, relevent_docs):
    from matplotlib import pyplot as plt
    # precision=0
    precision=[]
    recall=[]
    lent=len(relevent_docs)
    map_val=0
    i=0
    j=0
    for x in reterived_docs:
        j = j + 1
        if x in relevent_docs:
            i=i+1
        map_val=map_val+i/j
        
        precision.append(i/j)
        recall.append(i/lent)


    map_val=map_val/j
#     precision=i/j
#     recall=i/len(relevent_docs)
    plt.plot(recall,precision,  label="PR")
    plt.xlabel('Recall')
    # Labeling the Y-axis
    plt.ylabel('Precision')
    # Give a title to the graph
    plt.title('--PR--')

    # Show a legend on the plot
    plt.legend()

    plt.show()
    
    return map_val
```
